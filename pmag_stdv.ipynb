# Create a table of only pts with DAGE less than or equal to 1

In our range of 0-5 Ma

```python
AFstd[AFstd.DAGE<=1]
```

```python
Certain = AFstd[AFstd.DAGE<=1]
print(Certain.head())  # looking at the first few values of this filter
print(Certain.tail())  # looking at the last few values of this filter
```

```python
Certain.DAGE.max()
print('Maximum Stdv of Age:', Certain.DAGE.max())

Certain.DAGE.min()
print('Minimum Stdv of Age:', Certain.DAGE.min())
```

```python
DFMax = Certain.DF.max()
print('Maximum Stdv of Intensity',Certain.DF.max())

DFMin = Certain.DF.min()
print('Minimum stdv of Intensity',Certain.DF.min())
```

### Looking at the max and min of both Age and Intensity gives a range of this data.
We are mostly concerned with the max of the intensity to begin filtering those that are too large

```python
print('Average Stdv of Intensity {:0.2f}'.format(Certain.DF.mean()))
```

```python
# intensty stv points that are abnormally high

Peaks = AFstd[AFstd.DF>20]
print(Peaks.head())
```

```python
AFCertain = Certain[Certain.DF<20] 
AFCertain

# has DAGE<1  & DF<20
```
