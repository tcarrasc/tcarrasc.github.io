# PSV10 Residual Plots

### The Average residuals from the PSV10 data are plotted for average locations with a linear scaling factor. 

#### Read into the PSV10 Averages Dataframe from the PSV10_AvgGroups.ipynb file.

```python
import pandas as pd 

data = pd.read_pickle("PSV10_AvgGroups.pkl")

data["AvgLat"] = pd.to_numeric(data["AvgLat"]) #convert AvgLat values from dtype objects to float64 
data["AvgLong"] = pd.to_numeric(data["AvgLong"]) #convert AvgLong values from dtype objects to float64 

data
```

```python
data["ResidualWeighted"] = data["ΔIAvg"]/data["ΔIstdvWeighted"] 
data["Residual"] = data["ΔIAvg"]/data["ΔIstdv"] 
data
```

```python
data_ = data[data.GroupAmount != 64] # RW values of 7, 10, 14 
data_a = data_[data_.GroupAmount != 141]
data_b = data_a[data_a.GroupAmount != 39]
data_b
```

#### Round weighted residuals to a single decimal point. Create seperate tables for positive and negative residual values. 

#### For ease of plotting, create a new column to take the absolute value of the negative residuals. 

```python
import numpy as np 


data1 = data_b.copy()

data1["RW_rd"] = data1.ResidualWeighted  #copies the values from the ResidualsWeighted to a new column

# Rounds values in new RW_rd column to 1 decimal point
decimals = 1    
data1['RW_rd'] = data1['RW_rd'].apply(lambda x: round(x, decimals))

RW_pos = data1[data1.RW_rd >= 0] # new table only with data that has positive (greater than 0) residuals
RW_neg = data1[data1.RW_rd <= 0] # new table only with data that has negative (less than 0) residuals

RW_neg['RW_rd_abs'] = np.absolute(RW_neg.loc[:,('RW_rd')]) #takes abs value of negative residuals 


data2 = data.copy()

data2["R_rd"] = data2.Residual  #copies the values from the Residuals to a new column

# Rounds values in new R_rd column to 1 decimal point
decimals = 1    
data2['R_rd'] = data2['R_rd'].apply(lambda x: round(x, decimals))

R_pos = data2[data2.R_rd >= 0] # new table only with data that has positive (greater than 0) residuals
R_neg = data2[data2.R_rd <= 0] # new table only with data that has negative (less than 0) residuals

R_neg['R_rd_abs'] = np.absolute(R_neg['R_rd']) #takes abs value of negative residuals 
#R_pos
RW_neg


#updated_RW_neg = RW_neg.drop(['1223','1017']) # weighted residual value 14 and 10
#updated_RW_neg
```

#### Set the region of the plot to be slightly larger than data bounds

```python
import pygmt

region = [
    data.AvgLong.min() - 1,
    data.AvgLong.max() + 1,
    data.AvgLat.min() - 1,
    data.AvgLat.max() + 1,
]

print(region)
print(data.head())
```

#### Plot the average locations

```python
fig = pygmt.Figure()
fig.basemap(region=region, projection="M15c", frame=True)
fig.coast(land="black", water="skyblue")
fig.plot(x=data.AvgLong, y=data.AvgLat, style="c0.3c", color="white", pen="black")
fig.show()
```

#### Using the plus symbol, plot for the positive weighted residuals of the average locations. 

```python
import numpy as np
import matplotlib as plt

fig = pygmt.Figure()
fig.basemap(region=region, projection="M15c", frame =True)
fig.coast(land="white", water="skyblue")
fig.plot(
   x=RW_pos.AvgLong,
   y=RW_pos.AvgLat, 
   size=0.03 * (10*RW_pos.RW_rd),
   style="+cc",
   pen="1p,black",
)
fig.basemap(frame=["a", '+t" (+) Weighted Residuals of Avg locations"'])

fig.show()

#fig.savefig('../figs/' + 'residualscaling_positive.png', dpi=160)
```

#### Using the open circle symbol, plot for the negative weighted residuals of the average locations. 

```python
fig = pygmt.Figure()
fig.basemap(region=region, projection="M15c", frame=True)
fig.coast(land="white", water="skyblue")
fig.plot(
   x=RW_neg.AvgLong,
   y=RW_neg.AvgLat, 
   size=0.03 * (4*RW_neg.RW_rd_abs),
   style="cc",
   pen="1p,black",
)
fig.basemap(frame=["a", '+t"(-) Weighted Residuals of Avg locations"'])

fig.show()

#fig.savefig('../figs/' + 'residualscaling_negative.png', dpi=160)
```

#### Plot all the weighted residuals together 

```python
fig = pygmt.Figure()
fig.basemap(region=region, projection="M15c", frame=True)
fig.coast(land="white", water="skyblue")

fig.plot(
   x=RW_neg.AvgLong,
   y=RW_neg.AvgLat, 
   size=0.03 * (6*RW_neg.RW_rd_abs),
   style="cc",
   pen="1p,black") 

fig.plot(
   x=RW_pos.AvgLong,
   y=RW_pos.AvgLat, 
   size=0.03 * (6*RW_pos.RW_rd),
   style="+cc",
   pen="1p,black")

fig.basemap(frame=["a", '+t"Weighted Residuals of Avg locations"'])

fig.show()
fig.savefig('../figs/' + 'PSV10Wresidualscaling_all.png', dpi=160)
```

```python
fig = pygmt.Figure()
fig.basemap(region=region, projection="M15c", frame=True)
fig.coast(land="white", water="skyblue")

fig.plot(
   x=R_pos.AvgLong,
   y=R_pos.AvgLat, 
   size=0.03 * (6*R_pos.R_rd),
   style="+cc",
   pen="1p,black")

fig.plot(
   x=R_neg.AvgLong,
   y=R_neg.AvgLat, 
   size=0.03 * (6*R_neg.R_rd_abs),
   style="+cc",
   pen="1p,black")

fig.basemap(frame=["a", '+t"Residuals of Avg locations"'])

fig.show()
#fig.savefig('../figs/' + 'PSV10residualscaling_all.png', dpi=160)
```
